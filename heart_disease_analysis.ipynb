Welcome to this beginner-friendly data analysis project using a heart disease dataset.

We'll explore how different health features (like age, cholesterol, etc.) vary between people with and without heart disease.

This notebook is written to help **you** learn step-by-step using `pandas`, `matplotlib`, and `seaborn`.


```python
#Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look nicer
sns.set(style="whitegrid")
```

```python
#Load the dataset
dataset = pd.read_csv(r"c:\Users\abiol\Downloads\Week 7-Seminar 6-Predictive Analytics-20240429\Activity 1\heart_data.csv")  
```

```python
#View the first few rows
dataset.head()
```

```python
#Check structure and column types,displays column names, data types, and non-null counts
dataset.info()
```

```python
#Check for missing values which helps to see if any data is missing in each column
dataset.isnull().sum()
```

```python
#Count of heart disease presence,â€˜targetâ€™ is usually 1 = disease, 0 = no disease
dataset['target'].value_counts()
```

```python
#Visualize the count of heart disease presence
sns.countplot(x='target', data=dataset)
plt.title("Heart Disease Presence (1 = Yes, 0 = No)")
plt.show()
```

```python
#Explore how age is related to heart disease,and distributed for those with and without the disease
sns.histplot(data=dataset, x='age', hue='target', kde=True)
plt.title("Age Distribution by Heart Disease Status")
plt.show()
```

```python
#Comparing cholesterol levels by heart disease status, to spot if higher cholesterol is more common in one group
sns.boxplot(x='target', y='chol', data=dataset)
plt.title("Cholesterol Levels: With vs. Without Heart Disease")
plt.show()
```

```python
#Check correlations between numeric features using heatmap
#This helps to see if any features are strongly correlated with the target variable
plt.figure(figsize=(10, 6))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Heart Data")
plt.show()
```
